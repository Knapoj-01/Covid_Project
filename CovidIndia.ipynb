```python
import matplotlib.pyplot as plt
import pandas as pd
import Covid
```

```python
data = pd.read_csv('Data/India_2.csv')
a = Covid.SIR_Model(data, 14750000)
a.computeIC(14.5,100,2)
Ix1, Iy1 = a.predictI(1000,[1, 385])
fig, ax2 = plt.subplots()
ax2.plot(Ix1, Iy1)
ax2.scatter(a.getData("day")[1:380], a.getData("infecteds")[1:380], 0.8)
a.report()
```

```python
b = Covid.SIRD_Model(data, 6000000)
b.computeIC(15,5,65,2)
Ix1, Iy1 = b.predictI(1000,[1, 385])
c = Covid.SIRD_Model(data, 45000000)
c.computeIC(15,6,200,2)
Ix2, Iy2 = c.predictI(1000,[1, 385])
fig, ax2 = plt.subplots()
ax2.plot(Ix1, Iy1, Ix2,Iy2)
ax2.scatter(b.getData("day")[1:380], b.getData("infecteds")[1:380], 0.8)
b.report()
```

```python
#Initialize
d = Covid.SIRD_Model(data, 6000000)
#Initial Condition 1
d.computeIC(15,5,65,2)
Ix1, Iy1 = d.predictI(1000,[1, 130])
d.beginCont()
#Initial Condition 2
d.resetN(38000000)
d.computeIC(15,6,147,2)
Ix2, Iy2 = d.predictI(1000,[130, 192])
#Initial Condition 3
d.resetN(45500000)
d.computeIC(15,6,200,2)
Ix3, Iy3 = d.predictI(1000,[192, 275])
#Initial Condition 4
d.resetN(20000000)
d.computeIC(15,6,282,2)
Ix4, Iy4 = d.predictI(1000,[275, 360])
fig, ax2 = plt.subplots()
ax2.plot(Ix1, Iy1, Ix2,Iy2,Ix3, Iy3,Ix4, Iy4)
ax2.scatter(d.getData("day")[1:380], d.getData("infecteds")[1:380], 0.8)
print()
```
